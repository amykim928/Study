```python
#valid palindrome p138
import collections, re
class Solution:
    def isPalindrome(self, s: str) -> bool:
        strs = []
        for char in s:
            if char.isalnum(): #isalnum(): 영문자, 숫자 여부 판별 함수                
                strs.append(char.lower())
        while len(strs) > 1:
            if strs.pop(0) != strs.pop():
                return False
        return True

    def isPalindrome2(self, s: str) -> bool:
        strs: Deque = collections.deque()
        for char in s:
            if char.isalnum(): #isalnum(): 영문자, 숫자 여부 판별 함수                
                strs.append(char.lower())
        while len(strs) > 1:
            if strs.popleft()!= strs.pop():
                return False
        return True

    def isPalindrome3(self, s: str) -> bool:
        s = s.lower()
        s = re.sub('[^a-z0-9]','',s)
        return s == s[::-1]

sol = Solution()
s = "A man, a plan, a canal: Panama"
print(sol.isPalindrome(s))
s = "race a car"
print(sol.isPalindrome2(s))
s = "A man, a plan, a canal: Panama"
print(sol.isPalindrome3(s))
```

```python
#reverse string p145
class Solution:
    def reverse_string(self, s: list) -> list:
        s = s[::-1] #혹은 s.reverse()
        return s

    def reverse_string2(self, s: list) -> list:
        left, right = 0, len(s)-1
        while left < right:
            s[left], s[right] = s[right], s[left]
            left += 1
            right -= 1
        return s
        
sol = Solution()
a = ["H","a","n","n","a","h"]
print(sol.reverse_string(a))
print(sol.reverse_string2(a))
```

```python
#reorder log files p148
class Solution:
    def reorder_log_files(self, s: list) -> list:
        letters, digits = [], []
        for log in s:
            if log.split()[1].isdigit():
                digits.append(log)
            else:
                letters.append(log)
        letters.sort(key = lambda x: (x.split()[1:], x.split()[0]))
        #코드 설명: 첫번째 키로 우선 정렬한 후 동일하면 다음 키로 정렬
        return letters+digits

sol = Solution()
logs = ["dig1 8 1 5 1","let1 art can","dig2 3 6","let2 own kit dig","ket3 art zero"]
logs = sol.reorder_log_files(logs)
print(logs)
```

```python
#most common word p151
from collections import Counter
class Solution:
    def most_common_word(self, paragraph: str, banned: list) -> str:
        #words = [word for word in re.sub(r'[^\w]',' ',paragraph).lower().split() if word not in banned]
        words = []
        word = re.sub(r'[^\w]',' ',paragraph).lower().split()
        for w in word:
            if w not in banned:
                words.append(w)
        counts = collections.Counter(words)
        return counts.most_common(1)[0][0]

sol = Solution()
paragraph = "Bob hit a ball, the hit BALL flew far after it was hit."
banned = ["hit"]
print(sol.most_common_word(paragraph, banned))
```

```python
#group anagrams p153
import collections
class Solution:
    def group_anagrams(self, strs: list):
        anagrams = collections.defaultdict(list)
        for word in strs:
            anagrams[''.join(sorted(word))].append(word)
        return list(anagrams.values())

sol = Solution()
strs = ["eat","tea","tan","ate","nat","bat"]
a = sol.group_anagrams(strs)
print(a)
```

```python
#longest palindrome substring p159
class Solution:
    def longest_palindrome(self, s: str) -> str:
        long_pal = ""
        for i in range(0, len(s)):
            for j in range(i+1, len(s)+1):
                slice_str = s[i:j]
                if slice_str == slice_str[::-1] and len(slice_str)>len(long_pal):
                    long_pal = slice_str
        return long_pal

sol = Solution()
strs = "babad"
print(sol.longest_palindrome(strs))
```

```python
#two sum p173
class Solution:
    def two_sum(self, nums: list, target: int) -> list:
        nums_map = {}
        for i, num in enumerate(nums):
            nums_map[num] = i

        for i, num in enumerate(nums):
            if target - num in nums_map and i != nums_map[target - num]:
                return [i, nums_map[target - num]]

sol = Solution()
nums = [2, 7, 11, 15]
target = 9
print(sol.two_sum(nums, target))
```

```python
#3sum p184
class Solution:
    def three_sum(self, nums: list) -> list:
        result = []
        nums.sort()
        for i in range(len(nums)-2):
            if i>0 and nums[i] == nums[i-1]:
                continue
            left = i+1
            right = len(nums)-1
            #print(left, right)
            while left < right:
                sum = nums[i] + nums[left] + nums[right]
                if sum < 0:
                    left += 1
                elif sum > 0:
                    right -= 1
                else:
                    result.append([nums[i], nums[left], nums[right]])

                    while left < right and nums[left] == nums[left+1]:
                        left += 1
                    while left < right and nums[right] == nums[right-1]:
                        right -= 1
                    left += 1
                    right -= 1
        return result

sol = Solution()
nums = [-1, 0, 1, 2, -1, -4]

print(sol.three_sum(nums))
```

```python
#product of array except self p193
class Solution:
    def product_except_self(self, nums: list) -> list:
        result = []
        p = 1
        for i in range(0, len(nums)):
            result.append(p)
            p = p * nums[i]

        p = 1
        for i in range(len(nums)-1, 0 - 1, -1):
            result[i] *= p
            p = p * nums[i]
        return result

sol = Solution()
nums = [1, 2, 3, 4]
print(sol.product_except_self(nums))
```

```python
#palindrome linked list p201
import collections
class ListNode:
    def __init__(self, val = 0, next = None):
        self.val = val
        self.next = next
        
class Solution:
    def is_palindrome(self, head: ListNode) -> bool: #내 풀이 + 수정
        if not head:
            return True
        
        s = ""
        while head is not None:
            s += str(head.val)
            head = head.next
        return s == s[::-1]
    
    def is_palindrome2(self, head: ListNode) -> bool:
        q: List = []
        
        if not head:
            return True
        
        node = head
        
        while node is not None:
            q.append(node.val)
            node = node.next
        
        while len(q) > 1:
            if q.pop(0) != q.pop():
                return False
            
        return True
    
    def is_palindrome3(self, head: ListNode) -> bool:
        q: Deque = collection.deque()
        
        if not head:
            return True
        
        node = head
        
        while node is not None:
            q.append(node.val)
            node = node.next
        
        while len(q) > 1:
            if q.popleft() != q.pop():
                return False
            
        return True
    
    def is_palindrome4(self, head: ListNode) -> bool:
        rev = None
        slow = fast = head
        while fast and fast.next:
            fast = fast.next.next
            rev, rev.next, slow = slow, rev, slow.next
            
        if fast:
            slow = slow.next
            
        while rev and rev.val == slow.val:
            slow, rev = slow.next, rev.next
        return not rev
    
node1 = ListNode(1)
node2 = ListNode(2)
node1.next = node2

node = node1
sol = Solution()
print(sol.is_palindrome(node))
```

```python
#merge two sorted lists p213
class ListNode:
    def __init__(self, val = 0, next = None):
        self.val = val
        self.next = next
        
class Solution:
    def merge_two_list(self, l1: ListNode, l2: ListNode) -> ListNode:
        if not l1:
            return l2
        if not l2:
            return l1
        lis = []
        while l1 or l2:
            lis.append(l1.val)
            l1 = l1.next
            lis.append(l2.val)
            l2 = l2.next
        
        lis.sort()
        head = ListNode(lis[0])
        tail = head
        
        for i in range(1, len(lis)):
            tail.next = ListNode(lis[i])
            tail = tail.next
        
        return head
    
    def merge_two_list2(self, l1: ListNode, l2: ListNode) -> ListNode:
        if (not l1) or (l2 and l1.val > l2.val):
            l1, l2 = l2, l1
        if l1:
            l1.next = self.merge_two_list2(l1.next, l2)
        
        return l1
    
node1 = ListNode(1)
node2 = ListNode(2)
node3 = ListNode(4)
node1.next = node2
node2.next = node3
node4 = ListNode(1)
node5 = ListNode(3)
node6 = ListNode(4)
node4.next = node5
node5.next = node6

head, head2 = node1, node4

node = node1
sol = Solution()
result_node = sol.merge_two_list(head, head2)
result_node2 = sol.merge_two_list(head, head2)

while result_node is not None:
    print(result_node.val, "->")
    result_node = result_node.next

while result_node2 is not None:
    print(result_node2.val, "->")
    result_node2 = result_node2.next
```

```python
#valid parentheses p245
class Solution:
    def is_valid(self, s: str) -> bool:
        stack = []
        table = {')': '(', '}': '{', ']': '['}
        
        for char in s:
            if char not in table:
                stack.append(char)
            elif not stack or table[char] != stack.pop():
                return False
            
        return len(stack) == 0
s = "{()[]{}"
sol = Solution()
print(sol.is_valid(s))
```

```python
#daily temperatures p252
class Solution:
    def daily_temperatures(self, T: list) -> list:
        answer = [0] * len(T)
        stack = []
        for i, cur in enumerate(T):
            while stack and cur > T[stack[-1]]:
                last = stack.pop()
                answer[last] = i - last
            stack.append(i)
    
        return answer
                
T = [73, 74, 75, 71, 69, 72, 76, 73]
sol = Solution()
print(sol.daily_temperatures(T))
```

```python
#implement stack using queues p 255
import collections
class MyStack:

    def __init__(self):
        self.q = collections.deque()

    def push(self, x: int) -> None:
        self.q.append(x)
        for _ in range(len(self.q)-1):
            self.q.append(self.q.popleft())
        
    def pop(self) -> int:
        return self.q.popleft()        

    def top(self) -> int:
        return self.q[0]
        

    def empty(self) -> bool:
        return len(self.q) == 0
        
myStack = MyStack()
myStack.push(1)
myStack.push(2)
print(myStack.top())
print(myStack.pop())
print(myStack.empty())
```

```python
#implement queue using stack p257
class MyQueue:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.input = []
        self.output = []

    def push(self, x: int) -> None:
        """
        Push element x to the back of queue.
        """
        self.input.append(x)

    def pop(self) -> int:
        """
        Removes the element from in front of queue and returns that element.
        """
        self.peek()
        return self.output.pop()

    def peek(self) -> int:
        """
        Get the front element.
        """
        if not self.output:
            while self.input:
                self.output.append(self.input.pop())
        return self.output[-1]

    def empty(self) -> bool:
        """
        Returns whether the queue is empty.
        """
        return self.input == [] and self.output == []
    
myQueue = MyQueue();
myQueue.push(1)
myQueue.push(2)
print(myQueue.peek())
print(myQueue.pop())
print(myQueue.empty())
```
