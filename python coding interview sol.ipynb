```python
#reorder log files p148
class Solution:
    def reorder_log_files(self, s: list) -> list:
        letters, digits = [], []
        for log in s:
            if log.split()[1].isdigit():
                digits.append(log)
            else:
                letters.append(log)
        letters.sort(key = lambda x: (x.split()[1:], x.split()[0]))
        #코드 설명: 첫번째 키로 우선 정렬한 후 동일하면 다음 키로 정렬
        return letters+digits

sol = Solution()
logs = ["dig1 8 1 5 1","let1 art can","dig2 3 6","let2 own kit dig","ket3 art zero"]
logs = sol.reorder_log_files(logs)
print(logs)
```

```python
#two sum p173
class Solution:
    def two_sum(self, nums: list, target: int) -> list:
        nums_map = {}
        for i, num in enumerate(nums):
            nums_map[num] = i

        for i, num in enumerate(nums):
            if target - num in nums_map and i != nums_map[target - num]:
                return [i, nums_map[target - num]]

sol = Solution()
nums = [2, 7, 11, 15]
target = 9
print(sol.two_sum(nums, target))
```

```python
#3sum p184
class Solution:
    def three_sum(self, nums: list) -> list:
        result = []
        nums.sort()
        for i in range(len(nums)-2):
            if i>0 and nums[i] == nums[i-1]:
                continue
            left = i+1
            right = len(nums)-1
            #print(left, right)
            while left < right:
                sum = nums[i] + nums[left] + nums[right]
                if sum < 0:
                    left += 1
                elif sum > 0:
                    right -= 1
                else:
                    result.append([nums[i], nums[left], nums[right]])

                    while left < right and nums[left] == nums[left+1]:
                        left += 1
                    while left < right and nums[right] == nums[right-1]:
                        right -= 1
                    left += 1
                    right -= 1
        return result

sol = Solution()
nums = [-1, 0, 1, 2, -1, -4]

print(sol.three_sum(nums))
```

```python
#valid parentheses p245
class Solution:
    def is_valid(self, s: str) -> bool:
        stack = []
        table = {')': '(', '}': '{', ']': '['}
        
        for char in s:
            if char not in table:
                stack.append(char)
            elif not stack or table[char] != stack.pop():
                return False
            
        return len(stack) == 0
s = "{()[]{}"
sol = Solution()
print(sol.is_valid(s))
```

```python
#daily temperatures p252
class Solution:
    def daily_temperatures(self, T: list) -> list:
        answer = [0] * len(T)
        stack = []
        for i, cur in enumerate(T):
            while stack and cur > T[stack[-1]]:
                last = stack.pop()
                answer[last] = i - last
            stack.append(i)
    
        return answer
                
T = [73, 74, 75, 71, 69, 72, 76, 73]
sol = Solution()
print(sol.daily_temperatures(T))
```

```python
#implement stack using queues p 255
import collections
class MyStack:

    def __init__(self):
        self.q = collections.deque()

    def push(self, x: int) -> None:
        self.q.append(x)
        for _ in range(len(self.q)-1):
            self.q.append(self.q.popleft())
        
    def pop(self) -> int:
        return self.q.popleft()        

    def top(self) -> int:
        return self.q[0]
        

    def empty(self) -> bool:
        return len(self.q) == 0
        
myStack = MyStack()
myStack.push(1)
myStack.push(2)
print(myStack.top())
print(myStack.pop())
print(myStack.empty())
```

```python
#implement queue using stack p257
class MyQueue:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.input = []
        self.output = []

    def push(self, x: int) -> None:
        """
        Push element x to the back of queue.
        """
        self.input.append(x)

    def pop(self) -> int:
        """
        Removes the element from in front of queue and returns that element.
        """
        self.peek()
        return self.output.pop()

    def peek(self) -> int:
        """
        Get the front element.
        """
        if not self.output:
            while self.input:
                self.output.append(self.input.pop())
        return self.output[-1]

    def empty(self) -> bool:
        """
        Returns whether the queue is empty.
        """
        return self.input == [] and self.output == []
    
myQueue = MyQueue();
myQueue.push(1)
myQueue.push(2)
print(myQueue.peek())
print(myQueue.pop())
print(myQueue.empty())
```

```python
#design circular deque p268
class ListNode:
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right
        
class MyCircularDeque:

    def __init__(self, k: int):
        """
        Initialize your data structure here. Set the size of the deque to be k.
        """
        self.head, self.tail = ListNode(None), ListNode(None)
        self.maxlen, self.len = k,0
        self.head.right, self.tail.left = self.tail, self.head
        
    def _add(self, node: ListNode, new: ListNode):
        n = node.right
        node.right = new
        new.left, new.right = node, n
        n.left = new
        
    def _del(self, node: ListNode):
        n = node.right.right
        node.right = n
        n.left = node

    def insertFront(self, value: int) -> bool:
        """
        Adds an item at the front of Deque. Return true if the operation is successful.
        """
        if self.len == self.maxlen:
            return False
        self.len += 1
        self._add(self.head, ListNode(value))
        return True

    def insertLast(self, value: int) -> bool:
        """
        Adds an item at the rear of Deque. Return true if the operation is successful.
        """
        if self.len == self.maxlen:
            return False
        self.len += 1
        self._add(self.tail.left, ListNode(value))
        return True
        
    def deleteFront(self) -> bool:
        """
        Deletes an item from the front of Deque. Return true if the operation is successful.
        """
        if self.len == 0:
            return False
        self.len -= 1
        self._del(self.head)
        return True
        

    def deleteLast(self) -> bool:
        """
        Deletes an item from the rear of Deque. Return true if the operation is successful.
        """
        if self.len == 0:
            return False
        self.len -= 1
        self._del(self.tail.left.left)
        return True

    def getFront(self) -> int:
        """
        Get the front item from the deque.
        """
        return self.head.right.val if self.len else -1
        
    def getRear(self) -> int:
        """
        Get the last item from the deque.
        """
        return self.tail.left.val if self.len else -1
        

    def isEmpty(self) -> bool:
        """
        Checks whether the circular deque is empty or not.
        """
        return self.len == 0

    def isFull(self) -> bool:
        """
        Checks whether the circular deque is full or not.
        """
        return self.len == self.maxlen
        
circularDeque = MyCircularDeque(3)
print(circularDeque.insertLast(1))
print(circularDeque.insertLast(2))
print(circularDeque.insertFront(3))
print(circularDeque.insertFront(4))
print(circularDeque.getRear())
print(circularDeque.isFull())
print(circularDeque.deleteLast())
print(circularDeque.insertFront(4))
print(circularDeque.getFront())
```

```python
#permutations p341
import itertools
class Solution:
    def permute(self, nums: list) -> list:
        results= []
        prev_elements = []
        
        def dfs(elements):
            if len(elements) == 0:
                results.append(prev_elements[:])
            
            for e in elements:
                next_elements = elements[:]
                next_elements.remove(e)
                
                prev_elements.append(e)
                dfs(next_elements)
                prev_elements.pop()
            
        dfs(nums)
        return results
    
    def permute2(self, nums: list) -> list:
        return list(itertools.permutations(nums))
    
arr = [1, 2, 3]
sol = Solution()
print(sol.permute(arr))
print(sol.permute2(arr))
```

```python
#combination sum p351
class Solution:
    def combination_sum(self, candidates: list, target: int) -> list:
        results = []
        
        def dfs(csum, index, path):
            if csum < 0:
                return
            if csum == 0:
                results.append(path)
                return
            
            for i in range(index, len(candidates)):
                dfs(csum - candidates[i], i, path + [candidates[i]])
        
        dfs(target, 0, [])
        return results
        
candidate = [2, 3, 6, 7]
target = 7
sol = Solution()
print(sol.combination_sum(candidate, target))
```

```python
#network delay time p373
import collections, heapq
class Solution:
    def network_delay_time(self, times: list, N: int, K: int) -> int:
        graph = collections.defaultdict(list)
        #그래프 인접 리스트 구성
        for u, v, w in times:
            graph[u].append([v, w])
        
        #큐 변수: [[소요 시간, 정점]]
        Q = [(0, K)]
        
        dist = collections.defaultdict(int)
        
        while Q:
            time, node = heapq.heappop(Q)
            if node not in dist:
                dist[node] = time
                for v, w in graph[node]:
                    alt = time + w
                    heapq.heappush(Q, (alt, v))
                    
        if len(dist) == N:
            return max(dist.values())
        
        return -1
        
times = [[2, 1, 1], [2, 3, 1], [3, 4, 1]]
N = 4
K = 2
sol = Solution()
print(sol.network_delay_time(times, N, K))
```

```python
#cheapest flights within k steps p379
import collections, heapq
class Solution:
    def find_cheapest_price(self, n: int, flights: list, src: int, dst: int, K: int) -> int:
        graph = collections.defaultdict(list)
        for u, v, w in flights:
            graph[u].append((v, w))
        
        Q = [(0, src, K)]
        
        while Q:
            price, node, k = heapq.heappop(Q)
            if node == dst:
                return price
            if k >= 0:
                for v, w in graph[node]:
                    alt = price + w
                    heapq.heappush(Q, (alt, v, k-1))
                    
        return -1

sol = Solution()
print(sol.find_cheapest_price(3, [[0, 1, 100], [1, 2, 100], [0, 2, 500]], 0, 2, 0))
```

```python
#maximum depth of binary tree p387
import collections
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def max_depth(self, root: TreeNode) -> int:
        if root is None:
            return 0
        queue = collections.deque([root])
        depth = 0
        
        while queue:
            depth += 1
            for _ in range(len(queue)):
                cur_root = queue.popleft()
                if cur_root.left:
                    queue.append(cur_root.left)
                if cur_root.right:
                    queue.append(cur_root.right)
        return depth
        
root = TreeNode(3)

root.left = TreeNode(9)
root.right = TreeNode(20)
root.right.left = TreeNode(15)
root.right.right = TreeNode(7)

sol = Solution()
print(sol.max_depth(root))
```

```python
#invert binary tree p397
import collections
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def invert_tree(self, root: TreeNode) -> TreeNode:
        if root:
            root.left, root.right = self.invert_tree(root.right), self.invert_tree(root.left)
            return root
        return None
    
    def invert_tree2(self, root: TreeNode) -> TreeNode:
        queue = collections.deque([root])
        
        while queue:
            node = queue.popleft()
            if node:
                node.left, node.right = node.right, node.left
                queue.append(node.left)
                queue.append(node.right)
        
        return root
    
    def invert_tree3(self, root: TreeNode) -> TreeNode:
        stack = collections.deque([root])
        
        while stack:
            node = stack.pop()
            if node:
                node.left, node.right = node.right, node.left
                stack.append(node.left)
                stack.append(node.right)
                
        return root
    
    def invert_tree4(self, root: TreeNode) -> TreeNode:
        stack = collections.deque([root])
        
        while stack:
            node = stack.pop()
            if node:
                stack.append(node.left)
                stack.append(node.right)
                node.left, node.right = node.right, node.left
                
        return root
    

root = TreeNode(4)

root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(1)
root.left.right = TreeNode(3)
root.right.left = TreeNode(6)
root.right.right = TreeNode(9)

sol = Solution()
print(sol.invert_tree(root))
print(sol.invert_tree2(root))
print(sol.invert_tree3(root))
print(sol.invert_tree4(root))
```

```python
#balanced binary tree p413
import collections
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def is_balanced(self, root: TreeNode) -> bool:
        def check(root: TreeNode):
            if not root:
                return 0
            left = check(root.left)
            right = check(root.right)
            
            if left == -1 or right == -1 or abs(left - right) > 1:
                return -1
            return max(left, right) + 1
        
        return check(root) != -1
        
root = TreeNode(3)

root.left = TreeNode(9)
root.right = TreeNode(20)
root.right.left = TreeNode(15)
root.right.right = TreeNode(7)

sol = Solution()
print(sol.is_balanced(root))
```

```python
#minimum height trees p416
import collections
class Solution:
    def find_min_height_trees(self, n: int, edges: list) -> list:
        if n <= 1:
            return [0]
        
        graph = collections.defaultdict(list)
        for a, b in edges:
            graph[a].append(b)
            graph[b].append(a)
        leaves = []
        for i in range(n+1):
            if len(graph[i]) == 1:
                leaves.append(i)
                
        while n > 2:
            n -= len(leaves)
            new_leaves = []
            for leaf in leaves:
                neighbor = graph[leaf].pop()
                graph[neighbor].remove(leaf)
                
                if len(graph[neighbor]) == 1:
                    new_leaves.append(neighbor)
        
            leaves = new_leaves
        
        return leaves
        
n = 4
edges = [[1, 0], [1, 2], [1, 3]]
sol = Solution()
print(sol.find_min_height_trees(n, edges))
```

```python
#range sum of BST p431
import collections
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    val: int = 0
    def range_sum_BST(self, root: TreeNode, low: int, high: int) -> int:
        if root:
            self.range_sum_BST(root.right, low, high)
            if root.val >= low and root.val <=high:
                self.val += root.val
            self.range_sum_BST(root.left, low, high)
            
        return self.val
    
    def range_sum_BST2(self, root: TreeNode, low: int, high: int) -> int: #모든 방향을 탐색하는 브루트 포스 풀이
        if not root:
            return 0
        
        return (root.val if low <= root.val <= high else 0) + \
    self.range_sum_BST2(root.left, low, high) + \
    self.range_sum_BST2(root.right, low, high)
    
    def range_sum_BST3(self, root: TreeNode, low: int, high: int) -> int: #dfs로 불필요한 부분 가지치기
        def dfs(node: TreeNode):
            if not node:
                return 0
            if node.val < low:
                return dfs(node.right)
            elif node.val > high:
                return dfs(node.left)
            return node.val + dfs(node.left) + dfs(node.right)
        
        return dfs(root)
     
    def range_sum_BST4(self, root: TreeNode, low: int, high: int) -> int: #반복 구조 DFS로 필요한 노드 탐색
        stack, sum = [root], 0
        while stack:
            node = stack.pop()
            if node:
                if node.val > low:
                    stack.append(node.left)
                if node.val < high:
                    stack.append(node.right)
                if low <= node.val <= high:
                    sum += node.val
                    
        return sum
        
    def range_sum_BST5(self, root: TreeNode, low: int, high: int) -> int: #반복 구조 BFS로 필요한 노드 탐색
        stack, sum = [root], 0
        while stack:
            node = stack.pop(0)
            if node:
                if node:
                    if node.val > low:
                        stack.append(node.left)
                    if node.val < high:
                        stack.append(node.right)
                    if low <= node.val <= high:
                        sum += node.val
        return sum
        
root = TreeNode(10)

root.left = TreeNode(5)
root.left.left = TreeNode(3)
root.left.right = TreeNode(7)
root.right = TreeNode(15)
root.right.right = TreeNode(18)

low = 7
high = 15

sol = Solution()
print(sol.range_sum_BST(root, low, high))
print(sol.range_sum_BST2(root, low, high))
print(sol.range_sum_BST3(root, low, high))
print(sol.range_sum_BST4(root, low, high))
print(sol.range_sum_BST5(root, low, high))
```

```python
#minimum distance between BST nodes p435
import sys
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    prev = -sys.maxsize
    result: int = sys.maxsize
    def min_diffIn_BST(self, root: TreeNode) -> int: #재귀 구조로 중위 순회
        if root.left:
            self.min_diffIn_BST(root.left)
        
        self.result = min(self.result, root.val - self.prev)
        self.prev = root.val
        
        if root.right:
            self.min_diffIn_BST(root.right)
            
        return self.result
    
    def min_diffIn_BST2(self, root2: TreeNode) -> int: #반복 구조로 중위 순회
            prev2 = -sys.maxsize
            result2 = sys.maxsize
            
            stack = []
            node = root2
            
            while stack or node:
                while node:
                    stack.append(node)
                    node = node.left
                
                node = stack.pop()
                
                result2 = min(result2, node.val - prev2)
                prev2 = node.val
                
                node = node.right
                
            return result2
        

root = TreeNode(4)

root.left = TreeNode(2)
root.left.left = TreeNode(1)
root.left.right = TreeNode(3)
root.right = TreeNode(6)

sol = Solution()
print(sol.min_diffIn_BST(root))
print(sol.min_diffIn_BST2(root))
```

```python
#construct binary tree from preorder and inorder traversal p444
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def build_tree(self, preorder: list, inorder: list) -> TreeNode:
        if inorder:
            idx = inorder.index(preorder.pop(0))
        
            node = TreeNode(inorder[idx])
            node.left = self.build_tree(preorder, inorder[0:idx])
            node.right = self.build_tree(preorder, inorder[idx+1:])
    
            return node
        
def postorder(node: TreeNode):
    if node is None:
        return
    postorder(node.left)
    postorder(node.right)
    print(node.val)
    
preorder = [3, 9, 20, 15, 7]
inorder = [9, 3, 15, 20, 7]

sol = Solution()
result = sol.build_tree(preorder, inorder)
postorder(result)
```

```python
#Kth largest element in an array p456
import heapq
class Solution:
    def find_Kth_largest(self, nums: list, k: int) -> int: #heapq 모듈 사용
        heap = list()
        for n in nums:
            heapq.heappush(heap, -n)
        
        for _ in range(1, k):
            heapq.heappop(heap)
            
        return -heapq.heappop(heap)
    
    def find_Kth_largest2(self, nums: list, k: int) -> int: #heapq 모듈의 heapify 이용
        heapq.heapify(nums) #리스트를 힙 형태로
        
        for _ in range(len(nums) - k):
            heapq.heappop(nums)
            
        return heapq.heappop(nums)
    
    def find_Kth_largest3(self, nums: list, k: int) -> int: #heapq 모듈의 nlargest 사용
        return heapq.nlargest(k, nums)[-1] #nsmallest도 가능
    
    def find_Kth_largest4(self, nums: list, k: int) -> int: #정렬 이용
        return sorted(nums, reverse = True)[k-1]
        
nums = [3, 2, 3, 1, 2, 4, 5, 5, 6]
k = 4
sol = Solution()
print(sol.find_Kth_largest4(nums, k))
```

```python
#implement trie (prefix tree) p461\
import collections
class TrieNode: #트라이를 저장할 노드
    def __init__(self):
        self.word = False
        self.children = collections.defaultdict(TrieNode)
        
class Trie:

    def __init__(self):
        self.root = TrieNode()

    def insert(self, word: str) -> None:
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.word = True
        

    def search(self, word: str) -> bool: #단어 존재 여부
        node = self.root
        for char in word:
            node = node.children[char]
        return node.word
        
    def startsWith(self, prefix: str) -> bool: #문자열로 시작 단어 존재 여부 판별
        node = self.root
        for char in prefix:
            if char not in node.children:
                return False
            node = node.children[char]
        return True
        
        
trie = Trie()
trie.insert("apple")
print(trie.search("apple"))
print(trie.search("app"))
print(trie.startsWith("app"))
trie.insert("app")
print(trie.search("app"))
```

```python
#palindrome pairs p466
class Solution:
    def palindrome_pairs(self, words: list) -> list:
        def is_palindrome(word):
            return word[:] == word[::-1]
        
        output = []
        for i, word1 in enumerate(words):
            for j, word2 in enumerate(words):
                if i==j:
                    continue
                if is_palindrome(word1+word2):
                    output.append([i, j])
        return output
        
words = ["abcd", "dcba", "lls", "s", "sssll"]
sol = Solution()
print(sol.palindrome_pairs(words))
```

```python
#sort list p489
class ListNode:
     def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
        
class Solution:
    def sort_list(self, head: ListNode) -> ListNode:
        node = head
        l = list()
        
        while node:
            l.append(node.val)
            node = node.next
        
        l.sort()
        
        node = head
        
        for i in range(len(l)):
            node.val = l[i]
            node = node.next
            
        return head
    
    def merge_two_lists(self, l1: ListNode, l2: ListNode) -> ListNode: #두 정렬 리스트 병합
        if l1 and l2:
            if l1.val > l2.val:
                l1, l2 = l2, l1
            l1.next = self.merge_two_lists(l1.next, l2)
            
        return l1 or l2
    
    def sort_list2(self, head: ListNode) -> ListNode:
        if not (head and head.next):
            return head
        
        half, slow, fast = None, head, head #런너 기법 활용
        while fast and fast.next:
            half, slow, fast = slow, slow.next, fast.next.next
        half.next = None
            
        l1=self.sort_list2(head) #분할 재귀 호출
        l2=self.sort_list2(slow)
        
        return self.merge_two_lists(l1, l2)
            
sol = Solution()

node = ListNode(4)
node2 = ListNode(2)
node3 = ListNode(1)
node4 = ListNode(3)

node.next = node2
node2.next = node3
node3.next = node4

result = sol.sort_list2(node)

while result:
    print(result.val)
    result = result.next
```

```python
#largest number p504
class Solution:
    @staticmethod
    def to_swap(n1: int, n2: int) -> bool:
        return str(n1) + str(n2) < str(n2) + str(n1)
    def largest_number(self, nums: list) -> str:
        i = 1
        while i < len(nums):
            j = i
            while j > 0 and self.to_swap(nums[j-1], nums[j]):
                nums[j], nums[j-1] = nums[j-1], nums[j]
                j -= 1
            i += 1
            
        return str(int(''.join(map(str, nums))))        
        
sol = Solution()
nums = [10, 2]
print(sol.largest_number(nums))
```

```python
#sort colors p508
class Solution:
    def sort_colors(self, nums: list) -> None:
        red, white, blue = 0, 0, len(nums)
        
        while white < blue:
            if nums[white] < 1:
                nums[red], nums[white] = nums[white], nums[red]
                white += 1
                red += 1
            elif nums[white] > 1:
                blue -= 1
                nums[white], nums[blue] = nums[blue], nums[white]
            else:
                white += 1
                
sol = Solution()
nums = [2, 0, 2, 1, 1, 0]
sol.sort_colors(nums)
print(nums)
```

```python
#K closet points to origin p511
import heapq
class Solution:
    def k_closest(self, points: list, k: int) -> list:
        heap = []
        for (x, y) in points:
            dist = x ** 2 + y ** 2
            heapq.heappush(heap, (dist, x, y))
        result = []
        for _ in range(k):
            (dist, x, y) = heapq.heappop(heap)
            result.append((x, y))
        return result
    
sol = Solution()
points = [[1, 3], [-2, 2]]
k = 1
print(sol.k_closest(points, k))
```
