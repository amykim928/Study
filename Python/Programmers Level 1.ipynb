```python
#4) 탐욕법 - 체육복
def solution(n, lost, reserve): #내 풀이
    cloth = set(i+1 for i in range(n))
    
    for i in range(len(lost)):
        if lost[i] in reserve:
            reserve.remove(lost[i])
            lost[i] = 0
    
    for i in lost:
        if i == 0:
            continue
        cloth.remove(i)
    
    for i in reserve:
        if i-1 not in cloth and i-1 > 0:
            cloth.add(i-1)
        elif i+1 not in cloth and i+1 <= n:
            cloth.add(i+1)
    
    return len(cloth)

#다른 사람의 풀이 (출처: https://programmers.co.kr/learn/courses/30/lessons/42862/solution_groups?language=python3)

def solution(n, lost, reserve):
    _reserve = [r for r in reserve if r not in lost]
    _lost = [l for l in lost if l not in reserve]
    for r in _reserve:
        f = r - 1
        b = r + 1
        if f in _lost:
            _lost.remove(f)
        elif b in _lost:
            _lost.remove(b)
    return n - len(_lost)

#다른 변수를 사용하지 않고 매개변수만 사용하여 해결한 풀이다.
```

```python
#20) 위클리 챌린지 - 4주차
def solution(table, languages, preference):
    table_list = []
    total_score = []

    table.sort()

    for t in table:
        table_list.append(t.split())

    for i in table_list:
        score = 0
        sub = i[1:]
        for j in range(len(languages)):
            if languages[j] in sub:
                score += (len(sub)-sub.index(languages[j]))*preference[j]
        total_score.append(score)

    idx = total_score.index(max(total_score))

    return table_list[idx][0]
```

```python
#46) 연습문제 - 최대공약수와 최소공배수
def solution(n, m):
    x, y = min(n, m), max(n, m)
    while y:
        x, y = y, x % y
    
    return [x, (n*m)/x]

#math 라이브러리에서 gcd, lcm 함수로 최대공약수, 최소공백수 계산 가능
#최대공약수 구할 때 유클리드 호제법! (왼쪽이 더 작은 수여야 함)
```

```python
#55) 위클리 챌린지 - 6주차_복서 정렬하기
import collections
def solution(weights, head2head):
    result = collections.defaultdict(list)
    for i in range(len(head2head)):
        win, lose, heavy = 0, 0, 0
        for j in range(len(head2head[i])):
            if head2head[i][j] == 'W':
                win += 1
                if weights[i] < weights[j]: heavy += 1
            if head2head[i][j] == 'L': lose += 1
        if win + lose != 0:
            result[i+1].append(win/(win+lose))
        else:
            result[i+1].append(0)
        result[i+1].append(heavy)
        result[i+1].append(weights[i])
    
    print(result)
    result = sorted(result, key = lambda x: (result[x][0], result[x][1], result[x][2]), reverse = True)
    return result

#다른 사람의 풀이 (출처: https://programmers.co.kr/learn/courses/30/lessons/85002/solution_groups?language=python3)
def solution(weights, head2head):
    result = []
    l = len(weights)
    # 한 번에 정렬해서 풀어봅시다!
    ans = [[0 for _ in range(4)] for _ in range(l)] # 승률, 무거운복서 이긴횟수, 자기 몸무게, 번호(음수로)
    for i in range(l):
        ans[i][2] = weights[i]
        ans[i][3] = -(i+1)
        cnt = 0 # 판수
        for j in range(l):
            if head2head[i][j] == 'W':
                ans[i][0] += 1 # 일단 이김
                cnt += 1
                if weights[i] < weights[j]:
                    ans[i][1] += 1 # 무거운 복서 이김
            elif head2head[i][j] == 'L':
                cnt += 1 # 판수만 늘려준다
        if cnt == 0:
            ans[i][0] = 0
        else:
            ans[i][0] /= cnt
    ans.sort(reverse=True) # 역순으로 정렬하면 모든 조건이 한 번에 정렬된다

    for i in range(l):
        result.append(-ans[i][3])
    return result

#리스트를 사용하여 정렬하면 lambda로 여러 번 정렬할 필요가 없다.
#번호만 작은 것이 앞쪽으로 오도록 정렬해야 하므로 음수 처리한다.
```
