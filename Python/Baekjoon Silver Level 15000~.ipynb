```python
#실버4 17626) Four Squares
#다른 사람의 코드 참고 (출처: https://minhyeok-rithm.tistory.com/entry/BOJ-17626)
import math, sys
n = int(sys.stdin.readline()) 
dp = [0] * (n + 1) #dp 프로그래밍
dp[0], dp[1] = 0, 1

for i in range(2, n + 1):
    if i == int(math.sqrt(i)) ** 2:
        dp[i] = 1
    else:
        dp[i] = i
for i in range(2, n + 1):
    for j in range(1, int(math.sqrt(i)) + 1):
        dp[i] = min(dp[i], dp[j * j] + dp[i - j * j])
        
print(dp[n])
```

```python
#실버3 18111) 마인크래프트
#다른 사람 코드 참고 (출처: https://nam-ki-bok.github.io/baekjoon/Baek_Minecraft/)
import collections, sys
n, m, b = map(int, sys.stdin.readline().split())
table = []
for i in range(n):
    table.extend(list(map(int, sys.stdin.readline().split())))

sum_table, len_table = sum(table), n*m
table = dict(collections.Counter(table))
height, min_time = 0, 10000000000000000

for i in range(257):
    time = 0
    if len_table * i <= sum_table + b:
        for key in table:
            if key > i:
                time += (key-i)*2*table[key]
            elif i > key:
                time += (i-key)*table[key]
        if time <= min_time:
            min_time = time
            height = i
            
print(min_time, height)
```

```python
#실버2 18870) 좌표 압축
import collections, sys
input = sys.stdin.readline
n = int(input())
arr = list(map(int, input().split()))
dic = collections.defaultdict(list)
for i in range(n):
    dic[arr[i]].append(i)

s = sorted(dic.items())
result = [0]*n
for i in range(len(s)):
    index = s[i][1]
    for j in index:
        result[j] = i

for i in result: print(i, end = ' ')

------------------------------------------

import sys
input = sys.stdin.readline
n = int(input())
arr = list(map(int, input().split()))
s = set(arr)
s = sorted(list(s))
dic = dict()
for i in range(len(s)):
    dic[s[i]] = i
result = []
for i in arr:
    result.append(dic[i])
for i in result: print(i, end = ' ')
```

```python
#실버4 19866) 덱
import sys, collections
n = int(sys.stdin.readline())
arr = collections.deque()
for i in range(n):
    n = sys.stdin.readline().strip()
    if 'push_front' in n:
        num = n.split()
        arr.appendleft(int(num[1]))
    elif 'push_back' in n:
        num = n.split()
        arr.append(int(num[1]))
    elif n == 'pop_front':
        print(arr.popleft()) if arr else print(-1)
    elif n == 'pop_back':
        print(arr.pop()) if arr else print(-1)
    elif n == 'size':
        print(len(arr))
    elif n == 'empty':
        print(0) if arr else print(1)
    elif n == 'front':
        print(arr[0]) if arr else print(-1)
    else:
        print(arr[-1]) if arr else print(-1)
```
