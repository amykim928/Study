```python
#스택
class Node:
    def __init__(self, item, next):
        self.item = item
        self.next = next
    
class Stack:
    def __init__(self):
        self.last = None
    def push(self, item):
        self.last = Node(item, self.last)
    def pop(self):
        item = self.last.item
        self.last = self.last.next
        return item
    
stack = Stack()
stack.push(1)
stack.push(2)
stack.push(3)
stack.push(4)
stack.push(5)

for _ in range(5):
    print(stack.pop())
```

```python
#DFS

def recursive_dfs(v, discovered = []):
    graph = { 1: [2, 3, 4], 2: [5], 3: [5], 4: [], 5: [6, 7], 6: [], 7: [3], }
    discovered.append(v)
    for w in graph[v]:
        if w not in discovered:
            discovered = recursive_dfs(w, discovered)
    return discovered

def iterative_dfs(start_v):
    graph = { 1: [2, 3, 4], 2: [5], 3: [5], 4: [], 5: [6, 7], 6: [], 7: [3], }
    discovered = []
    stack = [start_v]

    while stack:
        v = stack.pop()
        if v not in discovered:
            discovered.append(v)
            for w in graph[v]:
                stack.append(w)
    return discovered

print(recursive_dfs(1))
print(iterative_dfs(1))
```

```python
#BFS
def iterative_bfs(start_v):
    graph = { 1: [2, 3, 4], 2: [5], 3: [5], 4: [], 5: [6, 7], 6: [], 7: [3], }
    discovered = [start_v]
    queue = [start_v]
    while queue:
        v = queue.pop(0)
        for w in graph[v]:
            if w not in discovered:
                discovered.append(w)
                queue.append(w)
    return discovered

print(iterative_bfs(1))
```

```python
#heap
class BinaryHeap(object):
    def __init__(self):
        self.items = [None]
    def __len__(self):
        return len(self.items) - 1
    def _percolate_up(self):
        i = len(self)
        parent = i // 2
        while parent > 0:
            if self.items[i] < self.items[parent]:
                self.items[parent], self.items[i] = self.items[i], self.items[parent]
            i = parent
            parent = i // 2
    def insert(self, k):
        self.items.append(k)
        self._percolate_up()
        
    def _percolate_down(self, idx):
        left = idx * 2
        right = idx * 2 + 1
        smallest = idx
        
        if left <= len(self) and self.items[left] < self.items[smallest]:
            smallest = left
            
        if right <= len(self) and self.items[right] < self.items[smallest]:
            smallest = right
        
        if smallest != idx:
            self.items[idx], self.items[smallest] = self,items[smallest], self.items[idx]
            self._percolate_down(smallest)
        
    def extract(self):
        extracted = self.items[1]
        self.items[1] = self.items[len(self)]
        self.items.pop()
        self._percolate_down(1)
        return extracted

heap = BinaryHeap()
heap.insert(4)
print(heap.extract())
```

```python
#버블 정렬: 시간복잡도 O(n^2), 구현 가능한 가장 느린 정렬 알고리즘
def bubblesort(A):
    for i in range(1, len(A)):
        for j in range(0, len(A)-1):
            if A[j] > A[j+1]:
                A[j], A[j+1] = A[j+1], A[j]
                
#퀵 정렬: 피벗을 기준으로 좌우를 나누는 알고리즘. 매우 빠르지고 효율적이지만 입력값에 따라 성능 편차가 심함
def quicksort(A, lo, hi):
    def partition(lo, hi): #맨 오른쪽을 피벗으로 정하는 방식
        pivot = A[hi]
        left = lo
        for right in range(lo, hi):
            if A[right] < pivot:
                A[left], A[right] = A[right], A[left]
                left += 1
        A[left], A[hi] = A[hi], A[left]
        return left
    
    if lo < hi:
        pivot = partition(lo, hi)
        quicksort(A, lo, pivot - 1)
        quicksort(A, pivot + 1, hi)
```
