```python
#실버2 5525) IOIOI
#다른 사람의 풀이 참고(출처: https://donghak-dev.tistory.com/27)
import sys
input = sys.stdin.readline
n = int(input())
m = int(input())
s = input()

answer = 0
pattern = 0
i = 1

while i < m-1:
    if s[i-1] == 'I' and s[i] == 'O' and s[i+1] == 'I':
        pattern += 1
        i += 1
        if pattern == n:
            pattern -= 1
            answer += 1
    else:
        pattern = 0
    i += 1
print(answer)
```

```python
#실버1 6064) 카잉 달력
import sys
input = sys.stdin.readline
def cal(M, N, x, y):
    while x <= M*N:
        if (x-y)%N == 0:
            return x
        x += M
    return -1
T = int(input())
for i in range(T):
    M, N, x, y = map(int, input().split())
    print(cal(M, N, x, y))
```

```python
#실버1 7576) 토마토
import collections, sys
input = sys.stdin.readline
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def bfs():
    while ripe:
        nx, ny = ripe.popleft()
        for i in range(4):
            posx = nx + dx[i]
            posy = ny + dy[i]
            if 0 <= posx < n and 0 <= posy < m and graph[posx][posy] == 0:
                ripe.append((posx, posy))
                graph[posx][posy] = graph[nx][ny] + 1
                
m, n = map(int, input().split())
graph = []
ripe = collections.deque()
notRipe, afterNotRipe = False, False
result = 0
for i in range(n):
    graph.append(list(map(int, input().split())))

for i in range(n):
    for j in range(m):
            if graph[i][j] == 1:
                ripe.append((i, j))
            if graph[i][j] == 0:
                notRipe = True
if notRipe:
    bfs()
    for i in range(n):
        for j in range(m):
            result = max(result, graph[i][j])
            if graph[i][j] == 0:
                afterNotRipe = True
    if afterNotRipe:
        print(-1)
    else:
        print(result-1)
else:
    print(0)
```

```python
#실버5 7568) 덩치
n = int(input())
arr = []
result = []
for i in range(n):
    w, h = map(int, input().split())
    arr.append((w, h))
    
for i in arr:
    count = 0
    for j in range(len(arr)):
        if i[0] < arr[j][0] and i[1] < arr[j][1]:
            count += 1
    result.append(count+1)

for i in result:
    print(i, end = ' ')
```

```python
#실버4 9012) 괄호
def check(i):
    stack = []
    for c in i:
        if c == '(':
            stack.append(c)
        else:
            if stack:
                stack.pop()
            else:
                return 0
    if stack:
        return 0
    else:
        return 1

n = int(input())
for i in range(n):
    s = input()
    if check(s):
        print('YES')
    else:
        print('NO')
```

```python
#실버2 9184) 신나는 함수 실행
import sys
input = sys.stdin.readline
def w(a, b, c):
    if a <= 0 or b <= 0 or c <= 0:
        return 1

    if a > 20 or b > 20 or c > 20:
        return w(20, 20, 20)
    
    if dp[a][b][c]:
        return dp[a][b][c]
    
    if a < b < c:
        dp[a][b][c] = w(a, b, c-1) + w(a, b-1, c-1) - w(a, b-1, c)
        return dp[a][b][c]

    dp[a][b][c] = w(a-1, b, c) + w(a-1, b-1, c) + w(a-1, b, c-1) - w(a-1, b-1, c-1)
    return dp[a][b][c]

dp = [[[0]*21 for _ in range(21)] for _ in range(21)]
while True:
    a, b, c = map(int, input().split())
    if a == -1 and b == -1 and c == -1:
        break
    print(f'w({a}, {b}, {c}) = {w(a, b, c)}')
```
