```python
#실버2 1260) DFS와 BFS
import collections, sys
input = sys.stdin.readline

def bfs(graph, v):
    visited = []
    queue = collections.deque([v])
    while queue:
        n = queue.popleft()
        if n not in visited:
            visited.append(n)
            check = list(graph[n]-set(visited))
            check.sort()
            queue += check
    return ' '.join(str(i) for i in  visited)

def dfs(graph, v):
    visited = []
    stack = [v]
    while stack:
        n = stack.pop()
        if n not in visited:
            visited.append(n)
            check = list(graph[n]-set(visited))
            check.sort(reverse=True)
            stack += check
    return ' '.join(str(i) for i in  visited)

n, m, v = map(int, input().split())
graph = collections.defaultdict(set)
for i in range(m):
    n1, n2 = map(int, input().split())
    graph[n1].add(n2)
    graph[n2].add(n1)
print(dfs(graph, v))
print(bfs(graph, v))
```

```python
#실버1 2667) 단지번호붙이기
import sys
input = sys.stdin.readline
def dfs(x, y):
    global count
    if 0 > x or x >= n or 0 > y or y >= n or graph[x][y] == 0:
        return
    count += 1
    graph[x][y] = 0
    dfs(x+1, y)
    dfs(x-1, y)
    dfs(x, y+1)
    dfs(x, y-1)

n = int(input())
graph, result = [], []
total = 0
for i in range(n):
    graph.append(list(map(int, list(input().strip()))))
for i in range(n):
    for j in range(n):
        count = 0
        if graph[i][j]:
            dfs(i, j)
            total += 1
            result.append(count)
print(total)
result.sort()
for i in result: print(i)
```

```python
#실버3 2805) 나무 자르기
n, m = map(int, input().split())
arr = list(map(int, input().split()))
start, end = 1, max(arr)
while start <= end:
    mid = (start+end)//2 #이진탐색
    count = 0
    count = sum([i-mid if i-mid>0 else 0 for i in arr])
    
    if count >= m:
        start = mid + 1
    else:
        end = mid - 1
print(end)
```

```python
#실버3 3036) 링
def gcm(x, y):
    while y:
        x, y = y, x%y
    return x

n = int(input())
arr = list(map(int, input().split()))
for i in range(1, len(arr)):
    g = gcm(arr[0], arr[i])
    print(arr[0]//g, arr[i]//g, sep='/')
```

```python
#실버2 4948) 베르트랑 공준
def check(n):
    for i in range(2, int(n**(1/2))+1):
        if n % i == 0:
            return 0
    return 1

dp = [0]*(2*123456+1)
dp[0], dp[1] = 0, 0
for i in range(2, 2*123456+1):
    dp[i] = dp[i-1] + check(i)
while True:
    n = int(input())
    if n == 0:
        break
    print(dp[2*n]-dp[n])
```

```python
#실버4 4949) 균형잡힌 세상
while True:
    s = input()
    arr = []
    if s == '.':
        break
    for i in s:
        if i == '(' or i == '[': 
            arr.append(i)
        elif i == ')':
            if arr:
                if arr[-1] == '(':
                    arr.pop()
                else:
                    break
            else:
                arr.append(i)
                break
        elif i == ']':
            if arr:
                if arr[-1] == '[':
                    arr.pop()
                else:
                    break
            else:
                arr.append(i)
                break
            
    if arr: print('no')
    else: print('yes')
```
