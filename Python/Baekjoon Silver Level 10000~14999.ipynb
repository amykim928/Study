```python
#실버4 10845) 큐
import sys, collections
n = int(sys.stdin.readline())
arr = collections.deque()
for i in range(n):
    n = sys.stdin.readline().strip()
    if 'push' in n:
        num = n.split()
        arr.append(int(num[1]))
    elif n == 'pop':
        print(arr.popleft()) if arr else print(-1)
    elif n == 'size':
        print(len(arr))
    elif n == 'empty':
        print(0) if arr else print(1)
    elif n == 'front':
        print(arr[0]) if arr else print(-1)
    else:
        print(arr[-1]) if arr else print(-1)
```

```python
#실버2 11053) 가장 긴 증가하는 부분 수열
#최장 증가 부분 수열 알고리즘(LIS)
#dp 프로그래밍 (시간복잡도 O(n^2))
n = int(input())
arr = list(map(int, input().split()))
dp = [1]*n
for i in range(n):
    for j in range(i):
        if arr[j] < arr[i]:
            dp[i] = max(dp[i], dp[j]+1)
print(max(dp))

#이분탐색 (시간복잡도 O(nlogn))
def bisect(n):
    result = 0
    left, right = 0, len(dp)
    
    while left <= right:
        mid = (left+right)//2
        if dp[mid] >= n:
            result = mid
            right = mid-1
        else:
            left = mid+1       
    return result

n = int(input())
arr = list(map(int, input().split()))
dp = [arr[0]]

for i in range(n):
    if dp[-1] < arr[i]:
        dp.append(arr[i])
    else:
        dp[bisect(arr[i])] = arr[i] #bisect_left()로도 가능
        
print(len(dp))
```

```python
#실버2 11055) 가장 큰 증가 부분 수열
n = int(input())
arr = list(map(int, input().split()))
dp = arr[:]
for i in range(n):
    for j in range(i):
        if arr[j] < arr[i]:
            dp[i] = max(dp[i], dp[j]+arr[i])
print(max(dp))
```

```python
#실버2 11722) 가장 긴 감소하는 부분 수열 
def bisect(n):
    result = 0
    left, right = 0, len(dp)
    while left <= right:
        mid = (left+right)//2
        if dp[mid] <= n:
            right = mid - 1
            result = mid
        else:
            left = mid + 1
    return result

n = int(input())
arr = list(map(int, input().split()))
dp = [arr[0]]
for i in range(1, n):
    if dp[-1] > arr[i]:
        dp.append(arr[i])
    else:
        dp[bisect(arr[i])] = arr[i]
print(len(dp))
```

```python
#실버5 11723) 집합
import sys
m = int(sys.stdin.readline())
bit = 0
for _ in range(m):
    i = sys.stdin.readline().split()
    if i[0] == 'add':
        bit |= (1 << int(i[1])-1)
    elif i[0] == 'remove':
        bit &= ~(1 << int(i[1])-1)
    elif i[0] == 'check':
        if bit & (1 << int(i[1])-1) == 0:
            print(0)
        else:
            print(1)
    elif i[0] == 'toggle':
        bit ^= (1 << int(i[1])-1)
    elif i[0] == 'all':
        bit = (1 << 20) - 1
    else:
        bit = 0
```

```python
#실버1 11403) 경로 찾기
import sys
input = sys.stdin.readline
n = int(input())
graph = [list(map(int, input().split())) for _ in range(n)]
visit = [0 for i in range(n)]
def dfs(start):
    for i in range(n):
        if visit[i] == 0 and graph[start][i] == 1:
            visit[i] = 1
            dfs(i)
        
for i in range(n):
    dfs(i)
    for j in range(n):
        if visit[j] == 1:
            print(1, end = ' ')
        else:
            print(0, end = ' ')
    print()
    visit = [0 for i in range(n)]
```

```python
#실버2 11724) 연결 요소의 개수
import collections, sys
input = sys.stdin.readline
n, m = map(int, input().split())
graph = collections.defaultdict(set)
for i in range(m):
    u, v = map(int, input().split())
    graph[u].add(v)
    graph[v].add(u)

check = set(i for i in range(1, n+1))
visited = set()
stack = [1]
result = 0

while check:   
    while stack:
        n = stack.pop() 
        if n not in visited:
            check.remove(n)
            visited.add(n)
            stack += list(graph[n] - visited)
    add = list(check- visited)
    if add:
        stack.append(add.pop())
    result += 1
print(result)
```

```python
#실버3 14501) 퇴사
n = int(input())
T, P = [], []
for i in range(n):
    t, p = map(int, input().split())
    T.append(t)
    P.append(p)
    
dp = [0]*(n+1)
for i in range(n-1, -1, -1):
    if T[i]+i > n:
        dp[i] = dp[i+1]
    else:
        dp[i] = max(dp[i+1], P[i] + dp[i+T[i]])
    print(dp)
print(dp[0])
```
