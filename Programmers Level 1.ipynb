```python
#3) 완전탐색 - 모의고사
def solution(answers):
    supo1 = [1, 2, 3, 4, 5]
    supo2 = [2, 1, 2, 3, 2, 4, 2, 5]
    supo3 = [3, 3, 1, 1, 2, 2, 4, 4, 5, 5]
    
    check = [0, 0, 0, 0]
    
    for i in range(len(answers)):
        if supo1[i%5] == answers[i]:
            check[1] += 1
        if supo2[i%8] == answers[i]:
            check[2] += 1
        if supo3[i%10] == answers[i]:
            check[3] += 1
    
    m = max(check)
    return list(i for i, v in enumerate(check) if v == m)
```

```python
#4) 탐욕법 - 체육복
def solution(n, lost, reserve): #내 풀이
    cloth = set(i+1 for i in range(n))
    
    for i in range(len(lost)):
        if lost[i] in reserve:
            reserve.remove(lost[i])
            lost[i] = 0
    
    for i in lost:
        if i == 0:
            continue
        cloth.remove(i)
    
    for i in reserve:
        if i-1 not in cloth and i-1 > 0:
            cloth.add(i-1)
        elif i+1 not in cloth and i+1 <= n:
            cloth.add(i+1)
    
    return len(cloth)

#다른 사람의 풀이 (출처: https://programmers.co.kr/learn/courses/30/lessons/42862/solution_groups?language=python3)

def solution(n, lost, reserve):
    _reserve = [r for r in reserve if r not in lost]
    _lost = [l for l in lost if l not in reserve]
    for r in _reserve:
        f = r - 1
        b = r + 1
        if f in _lost:
            _lost.remove(f)
        elif b in _lost:
            _lost.remove(b)
    return n - len(_lost)

#다른 변수를 사용하지 않고 매개변수만 사용하여 해결한 풀이다.
```

```python
#5) 2021 Dev-Matching - 로또의 최고 순위, 최저 순위
def solution(lottos, win_nums):
    rank = [6, 6, 5, 4, 3, 2, 1]
    
    count = lottos.count(0)
    correct = 0
    
    for i in lottos:
        if i in win_nums:
            correct += 1
    
    
    return rank[correct+count], rank[correct]
```

```python
#6) 2021 KAKAO BLIND RECRUITMENT 신규 아이디 신청
def solution(new_id):
    #1단계
    new_id = new_id.lower()
    
    #2단계
    answer = ''
    for word in new_id:
        if word.isalnum() or word in '-_.':
            answer += word
            
    #3단계
    while '..' in answer:
        answer = answer.replace('..','.')
        
    #4단계
    answer = answer[1:] if answer[0] == '.' and len(answer) > 1 else answer
    answer = answer[:-1] if answer[-1] == '.' else answer
    
    #5단계
    if answer == '':
        answer = 'a'
    
    #6단계:
    if len(answer) >= 16:
        answer = answer[0:15]
        
    answer = answer[:-1] if answer[-1] == '.' else answer
    
    #7단계
    if len(answer) <= 2:
        answer += answer[-1]*(3-len(answer))
        
    return answer

#isalnum() 문자열이 영어, 한글, 숫자일때
#연속된 문자가 2번 이상일 때 제거 방법은 while과 replace 사용
#문자열 길이에 따른 예외 처리 중요
```

```python
#10) 월간 코드 챌린지 시즌2 - 음양 더하기
def solution(absolutes, signs):
    for i in range(len(signs)):
        if not signs[i]:
            absolutes[i] *= -1
            
    return sum(absolutes)
```

```python
#12) Summer/Winter Coding (~2018) - 소수 만들기
import math
import itertools
def solution(nums):
    answer = 0
    def check(num):
        for i in range(2, int(math.sqrt(num))+1):
            if num % i == 0:
                return False
        return True
    
    result = list(itertools.combinations(nums, 3))
    
    for r in result:
        if check(sum(r)):
            answer += 1
        
    return answer

#itertools는 내장함수 라이브러리로 사용 가능! 조합이나 순열이 필요할 때는 꼭 사용하자
```
